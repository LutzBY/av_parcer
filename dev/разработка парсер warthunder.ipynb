```python
import pyperclip
import re
import pandas as pd
import os
from datetime import datetime
```

```python
# парсер
def parse_battle_stats():
    text = pyperclip.paste()
    if not text.strip():
        print("❌ Буфер обмена пуст. Скопируй статистику боя и запусти скрипт снова.")
        return None

    # --- Результат: Победа / Поражение ---
    result_match = re.search(r'(Победа|Поражение) в миссии', text)
    result = result_match.group(1) if result_match else "Неизвестно"

    # --- Название миссии ---
    mission_match = re.search(r'миссии\s+"([^"]+)"', text)
    mission = mission_match.group(1) if mission_match else "Неизвестно"

    # --- Итого: СЛ, СОИ (FRP), ОИ (RP) — только последнее вхождение ---
    total_matches = re.findall(r'Итого:\s*(\d+)\s*СЛ,\s*(\d+)\s*СОИ,\s*(\d+)\s*ОИ', text)
    if not total_matches:
        print("❌ Не удалось найти ни одного вхождения 'Итого'.")
        return None

    # Берём ПОСЛЕДНЕЕ вхождение (финальные итоги)
    last_match = total_matches[-1]
    total_sl = int(last_match[0])
    total_frp = int(last_match[1])  # Free Research Points
    total_rp = int(last_match[2])   # Research Points

    # --- Очки миссии ---
    mission_points = re.findall(r'(\d+)\s*очк(?:о|а|ов)\s*миссии', text)
    total_mission_points = sum(int(x) for x in mission_points)

    # --- Сессия ---
    session_match = re.search(r'Сессия:\s*([a-f0-9]+)', text)
    session_id = session_match.group(1) if session_match else None
    if not session_id:
        print("❌ Не удалось найти session_id.")
        return None

    # --- Активность (%) ---
    activity_match = re.search(r'Активность:\s*(\d+)%', text)
    activity_percent = int(activity_match.group(1)) if activity_match else None

    # --- Использованная техника ---
    # Ищем строки из разделов "Время активности" и "Время игры"
    # Пример: "    ИС-6           1084 + (ПА)1084 = 2168 СЛ"
    #         "    Т-54 (1947)    1100 + (ПА)1100 = 2200 СЛ"
    #         "    ИС-6           60%     2:34"
    #         "    Т-54 (1947)    100%    9:59"

    # --- Использованная техника ---
    vehicles = set()

    # Паттерн 1: "Время активности" — ищем текст до "123 + (ПА)"
    pattern_active = r'^\s*(.+?)\s+\d+\s*\+\s*$$ПА$$'
    # Паттерн 2: "Время игры" — ищем текст до "95% ... 4:51"
    pattern_game = r'^\s*(.+?)\s+\d+%.*?\d+:\d+'

    active_time_matches = re.findall(pattern_active, text, re.MULTILINE)
    game_time_matches = re.findall(pattern_game, text, re.MULTILINE)

    all_vehicles = active_time_matches + game_time_matches

    for v in all_vehicles:
        cleaned = re.sub(r'\s+', ' ', v.strip())
        # Исключаем ложные срабатывания (например, "Заработано", "Итого")
        if cleaned and not re.match(r'^[0-9\[\]"]', cleaned) and len(cleaned) > 1:
            vehicles.add(cleaned)

    vehicles = ", ".join(sorted(vehicles)) if vehicles else "Неизвестно"

    return {
        'session_id': session_id,
        'vehicles': vehicles,
        'total_sl': total_sl,
        'total_frp': total_frp,
        'total_rp': total_rp,
        'total_mission_points': total_mission_points,
        'result': result,
        'mission': mission,
        'activity_percent': activity_percent,
    }

def save_to_excel(data):
    file_path = r"C:\Users\lutzb\Desktop\wt_stats\data.xlsx"
    columns = [
        'session_id', 'vehicles', 'total_sl', 'total_frp', 'total_rp',
        'total_mission_points', 'result', 'mission', 'activity_percent'
    ]

    # Создаём новую строку
    new_row = [data[col] for col in columns]

    try:
        # Попробуем прочитать существующий файл
        df = pd.read_excel(file_path, engine='openpyxl')
        # Преобразуем в список словарей (или просто добавим строку)
        df_new = pd.DataFrame([data], columns=columns)
        df_combined = pd.concat([df, df_new], ignore_index=True)
    except FileNotFoundError:
        # Если файла нет — создаём с нуля
        df_combined = pd.DataFrame([data], columns=columns)
    except Exception as e:
        print(f"⚠️ Ошибка при чтении файла: {e}. Создаём новый.")
        df_combined = pd.DataFrame([data], columns=columns)

    # Сохраняем обратно в Excel
    df_combined.to_excel(file_path, index=False, engine='openpyxl')
    print(f"✅ Добавлено в Excel: {file_path}")

# === ЗАПУСК ===
if __name__ == "__main__":
    print("🔄 Обработка буфера обмена...")
    data = parse_battle_stats()
    if data:
        print("\n📋 Извлечено:")
        for k, v in data.items():
            print(f"  {k}: {v}")
        save_to_excel(data)
    else:
        print("❌ Обработка не удалась.")
```
