```python
import pyperclip
import re
import pandas as pd
import os
from datetime import datetime
```

```python
# –ø–∞—Ä—Å–µ—Ä
def parse_battle_stats():
    text = pyperclip.paste()
    if not text.strip():
        print("‚ùå –ë—É—Ñ–µ—Ä –æ–±–º–µ–Ω–∞ –ø—É—Å—Ç. –°–∫–æ–ø–∏—Ä—É–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –±–æ—è –∏ –∑–∞–ø—É—Å—Ç–∏ —Å–∫—Ä–∏–ø—Ç —Å–Ω–æ–≤–∞.")
        return None

    # --- –†–µ–∑—É–ª—å—Ç–∞—Ç: –ü–æ–±–µ–¥–∞ / –ü–æ—Ä–∞–∂–µ–Ω–∏–µ ---
    result_match = re.search(r'(–ü–æ–±–µ–¥–∞|–ü–æ—Ä–∞–∂–µ–Ω–∏–µ) –≤ –º–∏—Å—Å–∏–∏', text)
    result = result_match.group(1) if result_match else "–ù–µ–∏–∑–≤–µ—Å—Ç–Ω–æ"

    # --- –ù–∞–∑–≤–∞–Ω–∏–µ –º–∏—Å—Å–∏–∏ ---
    mission_match = re.search(r'–º–∏—Å—Å–∏–∏\s+"([^"]+)"', text)
    mission = mission_match.group(1) if mission_match else "–ù–µ–∏–∑–≤–µ—Å—Ç–Ω–æ"

    # --- –ò—Ç–æ–≥–æ: –°–õ, –°–û–ò (FRP), –û–ò (RP) ‚Äî —Ç–æ–ª—å–∫–æ –ø–æ—Å–ª–µ–¥–Ω–µ–µ –≤—Ö–æ–∂–¥–µ–Ω–∏–µ ---
    total_matches = re.findall(r'–ò—Ç–æ–≥–æ:\s*(\d+)\s*–°–õ,\s*(\d+)\s*–°–û–ò,\s*(\d+)\s*–û–ò', text)
    if not total_matches:
        print("‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –Ω–∞–π—Ç–∏ –Ω–∏ –æ–¥–Ω–æ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è '–ò—Ç–æ–≥–æ'.")
        return None

    # –ë–µ—Ä—ë–º –ü–û–°–õ–ï–î–ù–ï–ï –≤—Ö–æ–∂–¥–µ–Ω–∏–µ (—Ñ–∏–Ω–∞–ª—å–Ω—ã–µ –∏—Ç–æ–≥–∏)
    last_match = total_matches[-1]
    total_sl = int(last_match[0])
    total_frp = int(last_match[1])  # Free Research Points
    total_rp = int(last_match[2])   # Research Points

    # --- –û—á–∫–∏ –º–∏—Å—Å–∏–∏ ---
    mission_points = re.findall(r'(\d+)\s*–æ—á–∫(?:–æ|–∞|–æ–≤)\s*–º–∏—Å—Å–∏–∏', text)
    total_mission_points = sum(int(x) for x in mission_points)

    # --- –°–µ—Å—Å–∏—è ---
    session_match = re.search(r'–°–µ—Å—Å–∏—è:\s*([a-f0-9]+)', text)
    session_id = session_match.group(1) if session_match else None
    if not session_id:
        print("‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –Ω–∞–π—Ç–∏ session_id.")
        return None

    # --- –ê–∫—Ç–∏–≤–Ω–æ—Å—Ç—å (%) ---
    activity_match = re.search(r'–ê–∫—Ç–∏–≤–Ω–æ—Å—Ç—å:\s*(\d+)%', text)
    activity_percent = int(activity_match.group(1)) if activity_match else None

    # --- –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω–∞—è —Ç–µ—Ö–Ω–∏–∫–∞ ---
    # –ò—â–µ–º —Å—Ç—Ä–æ–∫–∏ –∏–∑ —Ä–∞–∑–¥–µ–ª–æ–≤ "–í—Ä–µ–º—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏" –∏ "–í—Ä–µ–º—è –∏–≥—Ä—ã"
    # –ü—Ä–∏–º–µ—Ä: "    –ò–°-6           1084 + (–ü–ê)1084 = 2168 –°–õ"
    #         "    –¢-54 (1947)    1100 + (–ü–ê)1100 = 2200 –°–õ"
    #         "    –ò–°-6           60%     2:34"
    #         "    –¢-54 (1947)    100%    9:59"

    # --- –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω–∞—è —Ç–µ—Ö–Ω–∏–∫–∞ ---
    vehicles = set()

    # –ü–∞—Ç—Ç–µ—Ä–Ω 1: "–í—Ä–µ–º—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏" ‚Äî –∏—â–µ–º —Ç–µ–∫—Å—Ç –¥–æ "123 + (–ü–ê)"
    pattern_active = r'^\s*(.+?)\s+\d+\s*\+\s*$$–ü–ê$$'
    # –ü–∞—Ç—Ç–µ—Ä–Ω 2: "–í—Ä–µ–º—è –∏–≥—Ä—ã" ‚Äî –∏—â–µ–º —Ç–µ–∫—Å—Ç –¥–æ "95% ... 4:51"
    pattern_game = r'^\s*(.+?)\s+\d+%.*?\d+:\d+'

    active_time_matches = re.findall(pattern_active, text, re.MULTILINE)
    game_time_matches = re.findall(pattern_game, text, re.MULTILINE)

    all_vehicles = active_time_matches + game_time_matches

    for v in all_vehicles:
        cleaned = re.sub(r'\s+', ' ', v.strip())
        # –ò—Å–∫–ª—é—á–∞–µ–º –ª–æ–∂–Ω—ã–µ —Å—Ä–∞–±–∞—Ç—ã–≤–∞–Ω–∏—è (–Ω–∞–ø—Ä–∏–º–µ—Ä, "–ó–∞—Ä–∞–±–æ—Ç–∞–Ω–æ", "–ò—Ç–æ–≥–æ")
        if cleaned and not re.match(r'^[0-9\[\]"]', cleaned) and len(cleaned) > 1:
            vehicles.add(cleaned)

    vehicles = ", ".join(sorted(vehicles)) if vehicles else "–ù–µ–∏–∑–≤–µ—Å—Ç–Ω–æ"

    return {
        'session_id': session_id,
        'vehicles': vehicles,
        'total_sl': total_sl,
        'total_frp': total_frp,
        'total_rp': total_rp,
        'total_mission_points': total_mission_points,
        'result': result,
        'mission': mission,
        'activity_percent': activity_percent,
    }

def save_to_excel(data):
    file_path = r"C:\Users\lutzb\Desktop\wt_stats\data.xlsx"
    columns = [
        'session_id', 'vehicles', 'total_sl', 'total_frp', 'total_rp',
        'total_mission_points', 'result', 'mission', 'activity_percent'
    ]

    # –°–æ–∑–¥–∞—ë–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É
    new_row = [data[col] for col in columns]

    try:
        # –ü–æ–ø—Ä–æ–±—É–µ–º –ø—Ä–æ—á–∏—Ç–∞—Ç—å —Å—É—â–µ—Å—Ç–≤—É—é—â–∏–π —Ñ–∞–π–ª
        df = pd.read_excel(file_path, engine='openpyxl')
        # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤–∞—Ä–µ–π (–∏–ª–∏ –ø—Ä–æ—Å—Ç–æ –¥–æ–±–∞–≤–∏–º —Å—Ç—Ä–æ–∫—É)
        df_new = pd.DataFrame([data], columns=columns)
        df_combined = pd.concat([df, df_new], ignore_index=True)
    except FileNotFoundError:
        # –ï—Å–ª–∏ —Ñ–∞–π–ª–∞ –Ω–µ—Ç ‚Äî —Å–æ–∑–¥–∞—ë–º —Å –Ω—É–ª—è
        df_combined = pd.DataFrame([data], columns=columns)
    except Exception as e:
        print(f"‚ö†Ô∏è –û—à–∏–±–∫–∞ –ø—Ä–∏ —á—Ç–µ–Ω–∏–∏ —Ñ–∞–π–ª–∞: {e}. –°–æ–∑–¥–∞—ë–º –Ω–æ–≤—ã–π.")
        df_combined = pd.DataFrame([data], columns=columns)

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –æ–±—Ä–∞—Ç–Ω–æ –≤ Excel
    df_combined.to_excel(file_path, index=False, engine='openpyxl')
    print(f"‚úÖ –î–æ–±–∞–≤–ª–µ–Ω–æ –≤ Excel: {file_path}")

# === –ó–ê–ü–£–°–ö ===
if __name__ == "__main__":
    print("üîÑ –û–±—Ä–∞–±–æ—Ç–∫–∞ –±—É—Ñ–µ—Ä–∞ –æ–±–º–µ–Ω–∞...")
    data = parse_battle_stats()
    if data:
        print("\nüìã –ò–∑–≤–ª–µ—á–µ–Ω–æ:")
        for k, v in data.items():
            print(f"  {k}: {v}")
        save_to_excel(data)
    else:
        print("‚ùå –û–±—Ä–∞–±–æ—Ç–∫–∞ –Ω–µ —É–¥–∞–ª–∞—Å—å.")
```
